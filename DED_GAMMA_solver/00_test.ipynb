```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def extract_points_from_kfile(filepath):
    with open(filepath, 'r') as f:
        lines = f.readlines()

    # 'nid, x, y, z' 헤더 찾기
    start = None
    for i, line in enumerate(lines):
        if 'nid' in line.lower() and 'x' in line.lower() and 'y' in line.lower():
            start = i + 1
            break

    if start is None:
        raise ValueError("Header line with 'nid, x, y, z' not found.")

    data = []
    for line in lines[start:]:
        if line.strip() == '' or line.strip().startswith('*'):
            break
        parts = line.strip().split()
        if len(parts) >= 4:
            try:
                x, y, z = float(parts[1]), float(parts[2]), float(parts[3])
                data.append([x, y, z])
            except ValueError:
                continue

    return np.array(data)

# 파일 경로
bound_path = "/home/ftk3187/github/GAMMA/DED_GAMMA_solver/examples/Lshape/Lshape_bound.k"
zigzag_path = "/home/ftk3187/github/GAMMA/DED_GAMMA_solver/examples/Lshape/Lshape_zigzag.k"

bound_points = extract_points_from_kfile(bound_path)
zigzag_points = extract_points_from_kfile(zigzag_path)

def plot_with_transparency(ax, points, main_color, title, elev=None, azim=None):
    z = points[:, 2]
    below_or_eq = points[z <= 0]
    above = points[z > 0]

    # 회색 점: 투명하게
    ax.scatter(below_or_eq[:, 0], below_or_eq[:, 1], below_or_eq[:, 2],
               s=1, color='gray', alpha=0.2, label='z ≤ 0', depthshade=True)

    # 색 점: 불투명
    ax.scatter(above[:, 0], above[:, 1], above[:, 2],
               s=3, color=main_color, alpha=1.0, label='z > 0', depthshade=False)

    ax.set_title(title)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    if elev is not None and azim is not None:
        ax.view_init(elev=elev, azim=azim)
    ax.legend()

# 플롯 구성
fig = plt.figure(figsize=(16, 10))

# Bound
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
plot_with_transparency(ax1, bound_points, main_color='blue', title="Bound - 3D View")

ax2 = fig.add_subplot(2, 2, 2, projection='3d')
plot_with_transparency(ax2, bound_points, main_color='blue', title="Bound - XZ View", elev=0, azim=0)

# Zigzag
ax3 = fig.add_subplot(2, 2, 3, projection='3d')
plot_with_transparency(ax3, zigzag_points, main_color='red', title="Zigzag - 3D View")

ax4 = fig.add_subplot(2, 2, 4, projection='3d')
plot_with_transparency(ax4, zigzag_points, main_color='red', title="Zigzag - XZ View", elev=0, azim=0)

plt.tight_layout()
plt.show()
```
